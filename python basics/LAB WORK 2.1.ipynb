```python
#Q-1
#EVEN AND ODD NUMBER CHACK

num = int(input("Enter number here: "))

if num % 2 == 0:
    print("Number is even")
else:
    print("Number is odd")
```

```python
#Q-2
#AGE CATEGORY

age = int(input("Enter your age here: "))

if age <= 12 and age >= 0:
    print("You are a child")
elif age <= 19 and age >= 13:
    print("You are a teenager")
elif age <= 59 and age >= 20:
    print("You are an adult")
elif age >= 60:
    print("You are a senior citizen")
else:
    print("You have entered age in negative value")
```

```python
#Q-3
#LARGEST NUMBER

A = int(input("Enter first number: "))
B = int(input("Enter second number: "))
C = int(input("Enter third number: "))

if A > B and A > C:
    print("A is the largest number")
elif B > A and B > C:
    print("B is the largest number")
else:
    print("C is the largest number")
```

```python
#Q-5
#while loop until zero
A = int(input())

while A != 0:
  A = int(input())
print("loop over")
```

```python
#Q-7
#PRINT ALL EVEN NUMBERS BETWEEN 1 TO 50
num = 2  # Start from the first even number

while num <= 50:
    print(num)  # Print the even number
    num += 2  # Increment by 2 to get the next even number
```
