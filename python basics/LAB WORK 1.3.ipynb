```python
#Q-1
#USE OF TYPE CASTING
choice_1 = input("please enter your age in numbers:")

print(type(choice_1))#HERE USER ENTERED AGE BUT THATS NOT AN INTEGER
#WE HAVE TO CONVERT

choice_1 = int(choice_1)
print(type(choice_1))#NOW IT IS AN INTEGER

#ON THE OTHER HAND WE CAN TAKE USER INPUT AS SPECIFIC DATA TYPE

choice_2 = int(input("please enter your age in numbers:"))
print(type(choice_2))
print(type(choice_1))
```

```python
#Q-2
#FLOAT TO INT

HEIGHT = input("Enter your height: ")

print(type(HEIGHT))

HEIGHT = float(HEIGHT)
print(type(HEIGHT), "this is converted to float")
```

```python
#Q-3
#PROGRAMME TO TAKE INPUT AS A BOLEAN AND CONVERT INTO INT

answer_the_question = bool(input("Are you indian:"))
print(type(answer_the_question))
answer_the_question = int(answer_the_question)
print(type(answer_the_question))
```
