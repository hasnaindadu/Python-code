```python
# STUDENT DATA ORGANIZER

title = "!! Welcome to Student Data Organizer !!"
title_2 = "!! Let's Store Student's Data Smoothly !!"

print(title.title().center(70, "^"))
print(title_2.title().center(70))

Student_details = {}
subjects = ("Maths, english, Science, gujarati")

while True:
    op = input('''\nPlease select an option:\n
1. Add Student\n2. Display All Students\n3. Update Student Information\n4. Delete Student\n5. Display Subjects Offered\n6. Exit\nEnter your choice: ''')
    match op:
        case '1':
            Student_id = input("Enter Student ID: ")
            Student_name = input("Enter Student Name: ")
            Age = input("Enter Student Age: ")
            Dob = input("Enter Student Date of Birth (dd/mm/yyyy): ")
            Grade = input("Enter Student Grade: ")
            Subject_selected = input("Enter Subjects (separate with ','): ").split(',')
            Student_details[Student_id] = {
                "Name": Student_name,
                "Age": Age,
                "Dob": Dob,
                "Grade": Grade,
                "Subjects": Subject_selected
            }
            print("Student Added Successfully!\n")
        case '2':
            if not Student_details:
                print("No student records found.\n")
            else:
                for Student_id, details in Student_details.items():
                    print(f"\nStudent ID: {Student_id}")
                    for key, value in details.items():
                        print(f"  {key}: {value}")
                    print("-" * 30)
        case '3':
            Student_id = input("Enter Student ID to update: ")
            if Student_id not in Student_details:
                print("Student ID not found!\n")
                continue
            Student_details[Student_id]["Name"] = input("Enter new name: ")
            Student_details[Student_id]["Age"] = input("Enter new age: ")
            Student_details[Student_id]["Dob"] = input("Enter new Date of Birth (dd/mm/yyyy): ")
            Student_details[Student_id]["Grade"] = input("Enter new grade: ")
            Student_details[Student_id]["Subjects"] = input("Enter new subjects (separate with ','): ")
            print("student information updated successfully!\n")
        case '4':
            Student_id = input("Enter Student ID to delete: ")
            if Student_id in Student_details:
                del Student_details[Student_id]
                print("Student deleted successfully!\n")
            else:
                print("Student id not found!\n")
        case '5':
            print(f"Subjects offered: {subjects}\n")
        case '6':
            print("exited Successfully!\n")
            break
        case _:
            print("Invalid Option! Plesse try again.\n")
```
