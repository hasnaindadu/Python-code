```python
#LOGIC BOX PROJECT

title = "--: welcome to the pattern generator and number analyzer :--"
print(title.title().center(150,"*"))

while True:
  user_choice = input('''Please Select an Option\n
1. Pattern Generator\n
2. Number Analyzer\n
3. Exit\n
Enter your choice: ''')
  match user_choice:

    case _ if user_choice == "1":
      while True:

        choice_2 = input('''Choose a pattern type\n
1.Right-angle Triangle\n
2.Triangle\n
3.Left-angle Triangl\n
Enter your choice''')

      #choice_3 = input("Enter the number of rows: ")

        if choice_2 == "1":
          choice_3 = input("Enter the number of rows: ")
          for i in range(1,int(choice_3)+1):
            print("*"*i)
          break

        elif choice_2 == "2":
          choice_3 = input("Enter the number of rows: ")
          for i in range(1,int(choice_3)+1):
            space = " " * (int(choice_3)-i)
            stars = "*" * (2 * i - 1)
            print(space+stars)
          break

        elif choice_2 == "3":
          choice_3 = input("Enter the number of rows: ")
          for i in range(1,int(choice_3)+1):
            print(" " * (int(choice_3)-i) + "*" * i)
          break

        else:
          print("You have selected an invalid option!!")
        break

    case _ if user_choice == "2":
      a = int(input("Please enter the first number of range:"))
      b = int(input("Please enter the second number of range:"))
      for i in range(a,(b+1)):
        if i % 2 == 0:
          print("Number",i, "is Even")
        else:
          print("Number",i, "is Odd")
      print(f'Sum of all numbers from {a} to {b} is: {sum(range(a,b+1))}')

    case _ if user_choice == "3":
      print("Exitted Successfully!,Have a good day!,Byy")
      break
    case _ :
      print("You have selected an invalid option!!,Please try again")
```
